```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib
import datetime
```

# Monitor in each room at home 

```python
df = pd.read_excel('dat/expambient.xlsx', sheet_name='air-monitor')
del df['Unnamed: 0']
```

```python
df.rename(columns={
    '更新時刻': 'time', '場所': 'place', '温度':'temp', '湿度':'humidity',
    'd3': 'watt', 'd4': 'irms', 'd5': 'vrms', 'd6' : 'power_factor'
    },inplace=True)
```

# Temperature

```python
plt.rcParams['figure.figsize'] = [15, 10]
df2 = df.pivot(index='time', columns='place', values='temp')
df2.plot()
df2.describe()
```

# Humidity

```python
df3 = df.pivot(index='time', columns='place', values='humidity')
df3.plot()
df3.describe()
```

# Power usage

```python
df = pd.read_excel('dat/expambient.xlsx', sheet_name='power-monitor')
del df['Unnamed: 0']
df.rename(columns={
    '更新時刻': 'time', '場所': 'place', '温度':'temp', '湿度':'humidity',
    'ワット': 'watt', '電流': 'irms', '電圧': 'vrms', '力率' : 'power_factor'
    },inplace=True)
```

```python
df.plot(x='time',y=['watt', 'irms', 'vrms'])
df[['watt', 'irms', 'vrms']].describe()
```
